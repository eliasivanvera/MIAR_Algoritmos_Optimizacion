```python
#Torres de Hanoi
def torres_hanoi(N, origen, destino, pivote):
  if N==1:
    print(f"Mover bloque desde {origen} a {destino}")
    return

  torres_hanoi(N-1, origen, pivote, destino)
  print(f"mover bloque desde {origen} a {destino}")
  torres_hanoi(N-1, pivote, destino, origen)
torres_hanoi(8,1,3,2)
```

```python
#DEVOLVER CAMBIO
def cambio_moneda(CANTIDAD,SISTEMA):
  print("Sistema: ")
  print(SISTEMA)
  SOLUCION=[0 for i in range(len(SISTEMA))] #Crear lista de ceros (Tantos como la diemsion de SISTEMA)
  VALOR_ACUMULADO=0

  for i in range(len(SISTEMA)):
    monedas=int((CANTIDAD-VALOR_ACUMULADO)/SISTEMA[i])
    SOLUCION[i]=monedas
    VALOR_ACUMULADO+=monedas*SISTEMA[i]
    if VALOR_ACUMULADO==CANTIDAD: break

  return SOLUCION

SISTEMA=[25,10,5,1]
cambio_moneda(54, SISTEMA)
```

```python
#Algoritmo de Fuerza Bruta
def puntos_mas_cercanos(puntos):
    min_dist= float('inf')
    punto1, punto2 = None, None

    n = len(puntos)
    for i in range(n):
        for j in range(i + 1, n):
            dist_actual = abs(puntos[i] - puntos[j])
            if dist_actual < min_dist:
                min_dist = dist_actual
                punto1, punto2 = puntos[i], puntos[j]

    return punto1, punto2
```

```python
# Ejemplo de uso:
puntos = [3403, 4537, 9089, 9746, 7259]
punto1, punto2 = puntos_mas_cercanos(puntos)
print("Intento por F. Bruta:", punto1, punto2)
```

```python
# Divide y venceras
def pares_mc_fb(pts):
    min_dist = float('inf')
    p1, p2 = None, None

    n = len(pts)
    for i in range(n):
        for j in range(i + 1, n):
            d_actual = abs(pts[i] - pts[j])
            if d_actual < min_dist:
                min_dist = d_actual
                p1, p2 = pts[i], pts[j]

    return p1, p2

def pares_mc_dv(pts):
    def pmc_rec(pts_ord):
        n = len(pts_ord)

        if n <= 3:
            return pares_mc_fb(pts_ord)

        medio = n // 2
        izq = pts_ord[:medio]
        der = pts_ord[medio:]

        p1_izq, p2_izq = pmc_rec(izq)
        p1_der, p2_der = pmc_rec(der)

        delta = min(abs(p1_izq - p2_izq), abs(p1_der - p2_der))
        p1, p2 = combinar_puntos(pts_ord, delta)

        return p1, p2

    def combinar_puntos(pts_ord, delta):
        n = len(pts_ord)
        medio = n // 2
        p_medio = pts_ord[medio]

        strip = [p for p in pts_ord if abs(p - p_medio) < delta]
        strip.sort()

        min_dist = delta
        p1, p2 = pts_ord[0], pts_ord[1]  # Valor predeterminado

        for i in range(len(strip)):
            for j in range(i + 1, len(strip)):
                if abs(strip[j] - strip[i]) < min_dist:
                    min_dist = abs(strip[j] - strip[i])
                    p1, p2 = strip[i], strip[j]

        return p1, p2

    pts_ord = sorted(pts)
    p1, p2 = pmc_rec(pts_ord)

    return p1, p2



```

```python
# Ejemplo de uso:
puntos = [3403, 4537, 9089, 9746, 7259, 1500, 6000, 8000]
punto1, punto2 = pares_mc_dv(puntos)
print("Segundo intento (Divide y Vencerás):", punto1, punto2)
```
